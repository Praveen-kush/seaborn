five common plots that can be created using the Seaborn library along with their typical uses:

Scatter Plot: Scatter plots are used to visualize the relationship between two continuous variables. They are helpful in identifying patterns, clusters, or correlations in the data.

Histogram: Histograms are useful for visualizing the distribution of a single numerical variable. They display the frequency of data points falling within different bins or intervals.

Bar Plot: Bar plots are effective for comparing the mean or aggregate values of different categories. They are commonly used to show comparisons between groups or to display categorical data.

Box Plot: Box plots, also known as box-and-whisker plots, provide a summary of the distribution of a continuous variable. They show the median, quartiles, and potential outliers in the data.

Heatmap: Heatmaps are used to visualize the matrix-like data where the values are represented as colors. They are often employed to explore correlations or patterns in large datasets, particularly in fields like genetics, finance, or image processing.

The Python code to load the "fmri" dataset using seaborn's 'load_dataset' function and plot a line plot using "timepoint" on the x-axis and "signal" on the y-axis for different events and regions.

```python
import seaborn as sns
import matplotlib.pyplot as plt
fmri_data = sns.load_dataset("fmri")
plt.figure(figsize=(10, 6))
sns.lineplot(data=fmri_data,x="timepoint",y="signal",hue="event",style="region",markers=True,dashes=False)
plt.title("FMRI signal over Time for different Event and regions")
plt.xlabel("Timepoint")
plt.ylabel("signal")
plt.legend(title="Event")
plt.show()
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
titanic_data = sns.load_dataset("titanic")
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(data=titanic_data, x='pclass', y='age')
plt.title('Age Distribution by Passenger class')
plt.xlabel('Passenger class')
plt.ylabel('Age')
plt.subplot(1, 2, 2)
sns.boxplot(data=titanic_data, x='pclass', y='fare')
plt.title('Fare Distribution by Passenger class')
plt.xlabel('Passenger class')
plt.ylabel('Fare')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
diamonds_data = sns.load_dataset("diamonds")
plt.figure(figsize=(10, 6))
sns.histplot(data = diamonds_data, x = 'price', hue = 'cut', kde = True)
plt.title('Distribution of Diamond Prices by Cut')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.legend(title = 'cut')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
flights_data = sns.load_dataset("flights")
flights_pivot = flights_data.pivot("month","year","passengers")
plt.figure(figsize=(10, 8))
sns.heatmap(flights_pivot, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Passenger Count by Month and Year")
plt.xlabel("Year")
plt.ylabel("Month")
plt.show()
```
